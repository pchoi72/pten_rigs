```python
active = rigs.pivot_table(index='date', columns='availability', values='count', aggfunc=np.sum)
active.columns.name = None
active.index.name = 'Date'
active.reset_index(inplace = True)
# print(active.columns)
# print('-'*50)
# print(active.dtypes)
# print('-'*50)
# print(type(active))
active
```

```python
temp = rigs.loc[rigs['availability'] == 'Operating']
temp
```

```python
region = temp.pivot_table(index='date', columns='region', values='count', aggfunc=np.sum)
region.columns.name = None
region.index.name = 'Date'
region.reset_index(inplace = True)
region
```

```python
di = pd.read_csv('enverus.csv')
di = di.T
di.columns = di.iloc[0]
di = di.drop(di.index[0])
di.reset_index(inplace=True)
di.columns.name = None
di = di.rename(columns={'index':'Date','ANADARKO BASIN':'Anadarko','APPALACHIA':'Appalachia',\
                        'DJ BASIN':'Rockies','GULF COAST BASIN (INCLUDING EAGLE FORD)':'South Texas',\
                        'PERMIAN':'West Texas','WILLISTON BASIN':'North Dakota','OTHER':'Other'})
di['Date']= pd.to_datetime(di['Date'])
di = di.drop(di.index[[0,1]])
di.reset_index(drop=True, inplace=True)
di
```

```python
region2 = region.copy()
region2['Anadarko'] = region2['East & North Texas'] + region2['Mid Continent']
region2 = region2[['Date','Anadarko','Appalachia','Rockies','South Texas','West Texas','North Dakota']]
region2['Total'] = region2['Anadarko'] + region2['Appalachia'] + region2['Rockies'] + region2['South Texas'] + region2['West Texas'] + region2['North Dakota']
region2
```

```python
di2 = di[['Date','Anadarko','Appalachia','Rockies','South Texas','West Texas','North Dakota','Total']]
di2
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=di2.Date, y=di2.Total, name="Industry Rig Count",\
               line=dict(color='royalblue', width=5, dash='dot')),
    secondary_y=False)

fig.update_yaxes(range=[820, 825],secondary_y=False)

fig.add_trace(
    go.Scatter(x=region2.Date, y=region2.Total, name="PTEN Rig Count",\
              line=dict(color='red', width=2)),
    secondary_y=True)

fig.update_yaxes(range=[118, 123],secondary_y=True)

# Add figure title
# fig.update_layout(title_text="<b>Industry vs. Patterson-UTI Energy Rig Count</b>")
fig.update_layout(
    title={
        'text': "<b>Industry vs. Patterson-UTI Energy Active Rig Count</b>",
        'y':0.85,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

# Move legend to bottom left
fig.update_layout(legend_orientation="h")

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="<b>Industry</b> Rig Count", secondary_y=False)
fig.update_yaxes(title_text="<b>PTEN</b> Rig Count", secondary_y=True)

fig.show()
```
